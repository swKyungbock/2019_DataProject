### 2-3. 파이 그래프 그리기

```python
import matplotlib.pyplot as plt
fruit=['사과','바나나','딸기','오렌지','포도']
result=[7,6,3,2,2]
plt.rcParams['font.family']='Malgun Gothic'
plt.title('좋아하는 선택 과일 결과')
plt.pie(result)
plt.show()
```

```python
import matplotlib.pyplot as plt
fruit=['사과','바나나','딸기','오렌지','포도']
result=[7,6,3,2,2]
sum=0
for i in result:
    sum=sum+i
print(sum)
plt.rcParams['font.family']='Malgun Gothic'
plt.rcParams['figure.figsize']=[6,6] #원의 모양을 정원으로 하기 위함
plt.title('좋아하는 선택 과일 결과')
explode_value=(0.1,0,0,0,0)
plt.pie(result, labels=fruit, autopct='%.1f%%',startangle=90,counterclock=False)
plt.show()
```

### 2-3. 산점도 그래프 그리기

```python
import matplotlib.pyplot as plt
x=[]
y=[]
size=[]
c=[]
colors=['r','g','b','c','m','k','y']
plt.title('랜덤 좌표, 랜덤 사이즈의 산점도 그래프')
import random
for i in range(100) :
    x.append(random.randint(1,10000))
    y.append(random.randint(1,10000))
    size.append(random.randint(1,1000))
    c.append(random.choice(colors))

plt.scatter(x,y,s=size,c=c,alpha=0.5)
plt.show()
```

## [실습1] 2차원 그래프 그리기

```python
import matplotlib.pyplot as plt
plt.rcParams['font.family']='Malgun Gothic' #한글 처리
plt.rcParams['axes.unicode_minus']=False #마이너스 기호(-) 문제 해결
x=[]
y=[]
for i in range(-100,100) :
    x.append(i)
    y.append(2*i**2)
plt.title('2차원 그래프 그리기')
plt.xlabel('x축')
plt.ylabel('y축')
plt.plot(x,y)
plt.show()
```

## [실습2] 키와 몸무게의 관계를 나타내는 산점도 그래프 그리기

```python
import matplotlib.pyplot as plt
import csv
plt.rcParams['font.family']='Malgun Gothic'
plt.title('키와 몸무게의 관계')
f=open('./data/height_weight.csv','r')
data=csv.reader(f)
height=[] #키 데이터
weight=[] #몸무게 데이터
size=[]
colors=['r','g','b','c','m','k','y']
next(data)
for row in data :
    height.append(float(row[0]))
    weight.append(float(row[1]))
import random
for i in range(100):
    size.append(random.randint(1,1000))
plt.scatter(height,weight,s=size,c=colors,alpha=0.5)
plt.xlabel('Height(cm)')
plt.ylabel('Weight(kg)')
plt.grid(True)
plt.show()
```

## [실습4] 운동 시작 전과 후의 윗몸 일으키기 횟수 변화 막대 그래프 그리기

```python
import matplotlib.pyplot as plt
import csv
f=open('./data/exercise_result.csv','r')
data=csv.reader(f)
next(data)
memberID=[]
before_ex=[]
after_ex=[]
for row in data :
    memberID.append(row[0])
    before_ex.append(int(row[1]))
    after_ex.append(int(row[2]))
dataCnt=len(memberID) #회원의 수
f.close()
index=[]
index2=[]
bar_width=0.4
for i in range(dataCnt) :
    index.append(i)
    index2.append(index[i]+bar_width)
plt.rcParams['font.family']='Malgun Gothic'
plt.title('운동 시작 전와 후의 윗몸 일으키기 횟수 변화')
plt.bar(index,before_ex,tick_label=memberID,color='c',label='before exercise',width=bar_width)
plt.bar(index2,after_ex,color='m',label='after exercise',width=bar_width)
plt.legend()
plt.show()
```

## #3.[project] 우리나라 지역별 인구밀도 scatter graph로 나타내기

```python
import matplotlib.pyplot as plt
import csv

city=[] #도시이름
lat=[]#위도(latitude)
lon=[]#경도(logitude)
pop_den=[]#2018 인구밀도
f=open('./data/pop_density.csv','r')
data=csv.reader(f)
next(data)
for row in data :
    city.append(row[0])
    lat.append(float(row[1]))
    lon.append(float(row[2]))
    pop_den.append(int(row[-1]))
# print(city)
# print(lat)
# print(lon)
# print(pop_den)
#색 지정
colors=['r','g','b','c','m','k','y']
# x축: 경도, y축 : 위도
plt.rcParams['font.family']='Malgun Gothic'
plt.title('지역별 인구 밀도(2018)')
plt.scatter(lon,lat,s=pop_den,c=colors,alpha=0.5)
plt.xlabel('경도')
plt.ylabel('위도')
for x,y,name in zip(lon, lat, city) :
    plt.text(x,y,name)
plt.show()
```

## #3.[project] 매년 내 생일의 기온 변화 그래프 그리기

```python
import csv
f=open('./data/seoul_temp.csv','r')
rdata=csv.reader(f)
highTemp=[]
lowTemp=[]
year=[]
for i in range(2002,2020):
    year.append(i)
for row in rdata :
    date=row[0]
    month=date[5:7]
    day=date[8:]
    if(month=='11' and day=='24') :
        highTemp.append(row[-1])
        lowTemp.append(row[2])
f.close()
import matplotlib.pyplot as plt
plt.plot(year,highTemp,label='high')
plt.plot(year,lowTemp,label='low')
plt.title('내 생일의 기온 변화(2002~2020년 11월 24일)')
plt.grid(True)
plt.legend()
plt.show()
```
